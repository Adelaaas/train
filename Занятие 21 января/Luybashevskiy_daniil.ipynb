```python
import pandas as pd
import math
```

```python
data = pd.read_csv('country_vaccinations.csv')
data
```

```python
print(data.shape)
print(data.isna().sum())
print(data.dtypes)
# Категориальные признаки: country, iso_code, vaccines, source_name, source_website
```

```python
j=0
for i in data.isna().sum():
    if i > data.shape[0]/2:
        data = data.drop(data.columns[j], axis=1)
    j += 1
data
```

```python
j=0
sum=0
n=0

for i in data.dtypes:
    if i=="float64":
        for num in data[data.columns[j]]:
            if not math.isnan(num):
                sum+=num
                n+=1
        data[data.columns[j]] = data[data.columns[j]].fillna(sum/n)
        sum=0
        n=0
    j+=1
    
    

data
```
